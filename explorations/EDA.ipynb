# Exploratory Data Analysis

```python
# kollar igenom datan
df = spark.table("marathos.bronze.results_raw")
display(df)
```

```python
# Lite mer om datan + printar schema
print(f"Number of rows: {df.count()}")
print(f"Number of columns: {len(df.columns)}")

print("\nColumn names:")
print(df.columns)

print("\nSchema:")
df.printSchema()
```

```python
# Visar fram en beskrivning av datan
display(df.describe())
```

## Null Values

```python
# Visar fram antal null värden
from pyspark.sql.functions import col, count, when

null_df = df.select([
    count(when(col(c).isNull(), c)).alias(c)
    for c in df.columns
])

display(null_df)
```

## Unique events

```python
# Visar fram antal unika värden
from pyspark.sql.functions import countDistinct

df.select(
    countDistinct("Event name").alias("unique_events")
).display()
```

## Age distribution

```python
# Delar upp datumet i år, månad och dag
display(
    df.groupBy("Athlete age category")
      .count()
      .orderBy("Athlete age category")
)
```

## Countries

```python
# Visar fram antal tävlande per land
from pyspark.sql.functions import desc

display(
    df.groupBy("Athlete country")
      .count()
      .orderBy(desc("count"))
)
```
